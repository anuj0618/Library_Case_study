#### Standardisation of Publish date

```
spark.conf.set("spark.sql.legacy.timeParserPolicy", "LEGACY")
```

```
#Converting all the date formats present in column publish_date to the standard format
df_publishDate_final = df_authors_final.withColumn("publish_date_updated", f.expr('''
case
    when to_date(publish_date, "dd MMM yyyy") is not null then to_date(publish_date, "dd MMM yyyy")
    when to_date(publish_date, "dd/MM/yyyy") is not null then to_date(publish_date, "dd/MM/yyyy")
    when to_date(publish_date, "MM/dd/yyyy") is not null then to_date(publish_date, "MM/dd/yyyy")
    when to_date(publish_date, "MM-dd-yyyy") is not null then to_date(publish_date, "MM-dd-yyyy")
    when to_date(publish_date, "MM.dd.yyyy") is not null then to_date(publish_date, "MM.dd.yyyy")
    when to_date(publish_date, "MM/dd/yy") is not null then to_date(publish_date, "MM/dd/yy")
    when to_date(publish_date, "MM-dd-yy") is not null then to_date(publish_date, "MM-dd-yy")
    when to_date(publish_date, "MM/yyyy") is not null then to_date(publish_date, "MM/yyyy")
    when to_date(publish_date, "MMMM d, yyyy") is not null then to_date(publish_date, "MMMM d, yyyy")
    when to_date(publish_date, "MMMM d,yy") is not null then to_date(publish_date, "MMMM d,yy")
    when to_date(publish_date, "MMMM yyyy") is not null then to_date(publish_date, "MMMM yyyy")
    when to_date(publish_date, "MMMM, yyyy") is not null then to_date(publish_date, "MMMM, yyyy")
    when to_date(publish_date, "yyyy") is not null then to_date(publish_date, "yyyy")
    else null
end
'''))
```

```
df_publishDate_final.createOrReplaceTempView('df_publishDate_final')
```

```
spark.sql('''select distinct publish_date_updated from df_publishDate_final''').show()
```

#### Fetch Publish Year

```
#Fetching Publish Year from standarsised publish date column
df_publish_year = spark.sql('''
  select *,
  case 
    when publish_date_updated is not null then year(publish_date_updated) 
    else null
  end as publish_year
  from df_publishDate_final
''')
```

#### Clean Up Of data

```
#CleanUp of data
clean_data_df = spark.sql('''
  select * from df_publish_year 
  where number_of_pages > 20
  and publish_year >= 1950
  and publish_date_updated is not null
  and key is not null
  and authors is not null
  and title is not null
''')
```

```
clean_data_df.createOrReplaceTempView('clean_data_df')
```

```
clean_data_df.count()
```

#### Get the book with the most pages

```
point4_query1 = spark.sql('''
  select distinct key as book_key, title as book_title, number_of_pages from
  (select *, dense_rank() over (order by number_of_pages desc) as dnk_rnk from clean_data_df)
  where dnk_rnk = 1
''').toPandas()
```

```
point4_query1
```

#### Find the top 5 genres with most books

```
#Using same logic as tried in data profiling to standardise the genre column values and adding them in another column
df_explode_genres = spark.sql('''select *, explode(genres) as genre from clean_data_df''')

df_genres = df_explode_genres.withColumn("genre", lower(f.regexp_replace(f.col("genre"), "\.$", "")))

df_genres.createOrReplaceTempView("df_genres")

#select book with most pages using dense_rank function
point4_query2 = spark.sql('''
  select * from (
  select rank() over (order by books desc) as genre_rank, genre, books from
  (select genre, count(distinct key) as books from df_genres group by 1)
  ) where genre_rank <=5
  order by genre_rank
''').toPandas()
```

```
point4_query2
```

```
#Exploding Co-authors values to get data at each co-author level
point4_query3_co_author = spark.sql('''select *, explode(co_authors) as author_key from clean_data_df''')
```

```
point4_query3_co_author.createOrReplaceTempView("point4_query3_co_author")
```

#### Retrieve the top 5 authors who (co-)authored the most books.

```
point4_query3 = spark.sql('''
  select author_rank, author_key, books from (
  select *, rank() over (order by books desc) as author_rank from
  (select author_key, count(distinct key) as books from point4_query3_co_author group by 1)
  )
  where author_rank <= 5
  order by author_rank
''').toPandas()
```

```
point4_query3
```

#### Per publish year, get the number of authors that published at least one book.

```
#Explode the authors column to aggregate data for each author
point4_query4_explode_authors = spark.sql('''select *, explode(authors) as author_key from clean_data_df''')
```

```
point4_query4_explode_authors.createOrReplaceTempView('point4_query4_explode_authors')

point4_query4 = spark.sql('''select publish_year, count(distinct author_key) as authors from
  (select *, author_key from point4_query4_explode_authors)
  group by 1
  order by 1
''').toPandas()
```

```
pd.set_option("display.max_rows", None)
point4_query4
```

#### Find the number of authors and number of books published per month for years between 1950 and 1970.

```
point4_query5 = spark.sql('''
  select date_format(publish_date_updated, "yyyy-MM") as publish_year_month, count(distinct author_key) as authors, count(distinct key) as books from
  (select *, author_key from point4_query4_explode_authors)
  where publish_year >= 1950 and publish_year <= 1970
  group by 1
  order by 1
''').toPandas()
```
